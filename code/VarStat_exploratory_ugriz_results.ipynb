```python
%matplotlib inline 
# Plot 

add = 'Mag' # Mag
# We make one such plot per filter ...
for filter in 'ugriz':
    #filter  = 'u'
    fig, ax = plt.subplots(1,2, figsize=(12,6))

    chi2max = 10
    mask_chi2 = varPatchAll[filter+'chi2DOF'].values < chi2max
    mask_n_points = (varPatchAll[filter+'N'].values > 60) & (varPatchAll[filter+'N'].values < 160)
    sigmin=0
    sigmax=5
    mask_sigma = (varPatchAll[filter+'sigmaFull'+add].values > sigmin) & (varPatchAll[filter+'sigmaFull'+add].values <sigmax)
    mask_total = mask_n_points & mask_chi2 & mask_sigma
    
    #LEFT 
    nObj = len(varPatchAll[filter+'N'][mask_total].values)
    chi2 = varPatchAll[filter+'chi2DOF'][mask_total].values
    mag = varPatchAll[filter+'psfMean_corr'][mask_total].values
    H, xedges, yedges = np.histogram2d(mag,chi2, bins=50)
    H = np.rot90(H)
    H = np.flipud(H)
    # Mask zeros
    Hmasked = np.ma.masked_where(H==0,H) # Mask pixels with a value of zeropltssssss
    # Plot 2D histogram using pcolor
    image= ax[0].pcolormesh(xedges,yedges,np.log10(Hmasked), cmap='jet')
    #ax.plot(np.linspace(0,4), np.linspace(0,4), ls='--', lw=4)
    ax[0].set_xlabel(filter+' psfMean', fontsize=15)
    ax[0].set_ylabel(r'$\chi^{2}_{DOF}('+filter+')$', fontsize=15)
    ax[0].set_title(' %d<N<%d, %d objects'%(60,160,nObj))
    ax[0].tick_params(axis='both', which='major', labelsize=15) 
    ax[0].set_xlim(16,24)
    colorbar = fig.colorbar(image, ax = ax[0],  orientation='vertical')
    colorbar.set_label(r'$\log_{10}(counts)$', fontsize=15)
    

    #RIGHT 
    nObj = len(varPatchAll[filter+'N'][mask_total].values)
    r = varPatchAll[filter+'sigmaFull'+add][mask_total].values / varPatchAll[filter+'muFull'+add][mask_total].values
    mag = varPatchAll[filter+'psfMean_corr'][mask_total].values
    H, xedges, yedges = np.histogram2d(mag,np.log10(r), bins=50)
    H = np.rot90(H)
    H = np.flipud(H)
    # Mask zeros
    Hmasked = np.ma.masked_where(H==0,H) # Mask pixels with a value of zeropltssssss
    # Plot 2D histogram using pcolor
    image = ax[1].pcolormesh(xedges,yedges,np.log10(Hmasked), cmap='jet')
    #ax.plot(np.linspace(0,4), np.linspace(0,4), ls='--', lw=4)
    ax[1].set_xlabel(filter+' psfMean', fontsize=15)
    ax[1].set_ylabel(r'$\log_{10}(\sigma_{Full}('+filter+')/ \mu_{Full}('+filter+'))['+add+']}$', fontsize=15)
    ax[1].set_title(r'$'+str(sigmin)+'<\sigma<'+str(sigmax)+add+'$', fontsize=15)
    ax[1].tick_params(axis='both', which='major', labelsize=15) 
    ax[1].set_xlim(16,24)
    ax[1].set_ylim(-4,0.5)
    colorbar = fig.colorbar(image, ax = ax[1],  orientation='vertical')
    colorbar.set_label(r'$\log_{10}(counts)$', fontsize=15)
    #fig.subplots_adjust(wspace=0.26)
    plt.tight_layout()
    name = '/astro/users/suberlak/S13Agg_analysis/data_products/'+filter+'NCSA_mag_chi2_ratio_sig'+add+'_set_lims.png'
    print('Saved as %s '%name)
    plt.savefig(name)
    plt.show()
```
